```python
mean_points = np.mean(data['Points'])
mean_scores = np.mean(data['Score'])
mean_weights = np.mean(data['Weight'])
```

```python
median_points = np.median(data['Points'])
median_scores = np.median(data['Score'])
median_weights = np.median(data['Weight'])
```

```python
mode_points = stats.mode(data['Points']).mode[0]
mode_scores = stats.mode(data['Score']).mode[0]
mode_weights = stats.mode(data['Weight']).mode[0]
```

```python
variance_points = np.var(data['Points'])
variance_scores = np.var(data['Score'])
variance_weights = np.var(data['Weight'])
```

```python
std_deviation_points = np.std(data['Points'])
std_deviation_scores = np.std(data['Score'])
std_deviation_weights = np.std(data['Weight'])
```

```python
range_points = np.ptp(data['Points'])
range_scores = np.ptp(data['Score'])
range_weights = np.ptp(data['Weight'])
```
